```python
import matplotlib.pyplot as plt
import numpy as np
plt.suptitle("Graph")

x=np.array([1,2])
plt.subplot(3,3,1)
plt.title("Title 1")
plt.plot(x,'o')



x=np.array([2,4,2,4,2])
y=np.array([4,2,4,2,4])
plt.subplot(3,3,2)
plt.plot(x,marker='*',linewidth=2.5)
plt.plot(y,'-.',linewidth=2.5)




y=np.array([3,8,1,10])
plt.subplot(3,3,3)
plt.plot(y,marker='o',mfc='y',mec='k')


y=np.array([1,2,3,4,5])
x=np.array([1,2,3,4,5])
plt.subplot(3,3,4)
plt.title("Title 2")
plt.plot(y,x)
plt.grid()



x = np.array([5,7,8,7,2,17,2,9,4])
y = np.array([99,86,87,88,111,86,103,87,94])
sizes = np.array([20,50,100,200,500,1000,60,90,10,])
plt.subplot(3,3,5)
plt.scatter(x,y,s=sizes,alpha=0.5)


x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.subplot(3,3,6)
plt.bar(x,y,width=0.1,color='r')
plt.show()

y=np.array([35,20,20,25])
label=["BIKE","CAR","BUS","SHIP"]
myexplode=[0.2,0,0,0]
color=["black",'r','y','c']

plt.pie(y,startangle=90,explode=myexplode,colors=color,labels=label,shadow=True)
plt.legend()


plt.show()



```
